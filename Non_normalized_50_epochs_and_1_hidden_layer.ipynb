```python
import numpy as np
import pandas as pd
import keras as kr
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(predictors.shape[1],)))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model=regression_model()
sqerror=[]
for _ in range(50):
  out=model.fit(predictors, target, validation_split=0.3, epochs=50)
  sqerror.append(out.history['loss'][-1])
```

```python
print('The mean of the mean squared errors of 50 runs of a model with 1 hidden layer with 50 epochs is ', np.mean(sqerror))
print('The standard deviation of the mean squared errors of 50 runs of a model with 1 hidden layer with 50 epochs is ', np.std(sqerror))
```
